```python
data_train = pd.read_csv('house_price_train.csv')
data_test = pd.read_csv('house_price_test.csv')
```

```python
data_train.head()
```

```python
def load_data(data):
    y = data['price']
    # Chuyển y về dạng ma trận numpy và chuyển vị
    y = np.array([y]).T
    # Lấy đơn vị của giá nhà là (triệu USD)
    y = y / 10**6
    # Lấy tên của đặc trưng
    choose_features = [
        'bedrooms',
        'bathrooms',
        'floors',
        'view',
        'condition',
        'grade',
        'yr_built',
        'sqft_living',
        'sqft_lot',
        'sqft_above',
        'sqft_basement',
    ]

    # Lấy ra 2 cột đặc trưng vừa được chọn
    X = data[choose_features]
    # Chuyển về ma trận numpy
    X = np.array(X)
    # Lấy đơn vị là hàng triệu
    X = X / 10**6
    # Xây dựng X mở rộng
    # Tạo vector toàn số 1 có độ dài bằng số hàng của X, để thêm vào X
    one = np.ones((X.shape[0], 1))

    # Hàm hstack để nối vector vào
    # A = [[-3], [2]]
    # B = [[1] , [1]]
    # A hstack B => [[-3 1] [2 1]]
    X = np.hstack((one, X))
    return X, y
```

```python
X, y = load_data(data_train)
```

```python
XtX = X.T @ X
XtX_inv = np.linalg.inv(XtX)
Xty = X.T @ y
theta = XtX_inv @ Xty
theta = theta[:, 0]
print(np.linalg.det(XtX))
```

Vì vậy ý tưởng pp ma trận là 1 ý tưởng tồi.Chúng ta sẽ thử so sánh với thư viện Scikit-Learn xem coi đi đúng hướng không.

```python
#thử coi xài thư viện sklearn xem giải làm sao
regr = LinearRegression(fit_intercept=False).fit(X, y)
theta_sklearn = regr.coef_[0]
print('theta tính bằng công thức : ', theta)
print('theta tính bằng thư viện Scikit-Learn: ', theta_sklearn)
```

Lost function hồi quy tuyến tính n biến: 

```python
def loss_multivariate(X, y, theta):
    theta = theta.reshape(-1, 1)
    m = len(X)
    # theories:
    h = X @ theta
    loss = 1/(2*m) * np.sum((y - h) ** 2)
    return loss
```

so sánh lost function:

```python
loss_multivariate(X, y, theta)
```

```python

loss_multivariate(X, y, theta_sklearn)
```

# ***Ý tường mình tìm ra:*** Đạo hàm cực đoan để tìm ra cực tiểu: Gradient Descent. Ý tưởng chung là làm sao cho bài toán ngày càng nhỏ lại và ngày càng hội tụ tại vị trí cực tiểu. Ý tưởng là vậy.

```python
def dL(X, y, theta):
    theta = theta.reshape(-1, 1)
    m = len(X)
    return -(1/m) * np.sum(X * (y - X @ theta), axis=0)
```

```python
def gradient_descent():
    m, n = X.shape
    # Khởi tạo theta ngẫu nhiên
    theta = np.random.randn(n)
    # Chọn các tham số như số lần lặp và hệ số alpha
    iterations = 1000001
    alpha = 0.5
    for i in range(iterations):
        # Cập nhật theta theo công thức của GD
        theta = theta - alpha * dL(X, y, theta)
        # Tính hàm mất mát
        loss = loss_multivariate(X, y, theta)
        if i % 20000 == 0:
            # Xuất giá trị mất mát ra để theo dõi
            print('Iter {}. Loss = {}'.format(i, loss))
    return theta
```

```python
theta_gd = gradient_descent()
```
